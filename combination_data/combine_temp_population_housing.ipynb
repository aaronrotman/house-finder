```python
import pandas as pd
```

```python
population = pd.read_csv('../population/population_categories.csv')
population
```

```python
# population['population'] = population['population'].str.replace(',', '')
```

```python
population.loc[:,'population'] = population.loc[:, 'population'].astype(int)
```

```python
population_sorted = population.sort_values(by=['state', 'city'])
population_sorted
```

```python
temp_data = pd.read_csv('../temperature/original_files/redfin_temps_complete.csv')
temp_data
```

```python
temp_data_sorted = temp_data.sort_values(by=['temp_state', 'temp_city'])
temp_data_sorted
```

```python
temp_data_sorted.dtypes
```

```python

temp_data_sorted.loc[:,'temp_city'] = temp_data_sorted.loc[:, 'temp_city'].astype("string")
temp_data_sorted.loc[:,'temp_state'] = temp_data_sorted.loc[:, 'temp_state'].astype("string")
temp_data_sorted.loc[:,'temp_abbr'] = temp_data_sorted.loc[:, 'temp_abbr'].astype("string")


```

```python
temp_data_sorted.dtypes
```

```python
pop_temp_merge = temp_data_sorted.merge(population_sorted, how="left", left_on=['temp_state', 'temp_city'], right_on=['state', 'city'])
```

```python
pop_temp_merge
```

```python
pop_temp_drop_col = pop_temp_merge.drop(columns=['state', 'city'])
```

```python
pop_temp_drop_col
```

## JOINING HOUSING

```python
redfin_data = pd.read_csv('../redfin_housing/redfin_resources/combined_regions/redfin_all_regions.csv')
```

```python
redfin_data
```

```python
redfin_sort_state = redfin_data.sort_values(by=['state', 'city'])
redfin_sort_state
```

```python
redfin_sort_dropna = redfin_sort_state.dropna()
redfin_sort_dropna
```

```python
pop_temp_drop_col.sort_values(by=['temp_state', 'temp_city'])
pop_temp_drop_col
```

```python
housing_complete_merge = redfin_sort_dropna.merge(pop_temp_drop_col, how="left", left_on=['state', 'city'], right_on=['temp_abbr', 'temp_city'])
housing_complete_merge
```

```python
housing_complete_merge['house_id'] = housing_complete_merge.index + 1
```

```python
housing_complete_merge
```

```python
housing_complete_merge.to_csv("housing_complete_for_cleaning_only.csv")
```

```python
cols = list(housing_complete_merge.columns.values)
cols
```

```python
complete_housing_data = housing_complete_merge[['house_id','property_type','address', 'city','state','zip_code',
                                                'price','beds','baths','square_feet','lot_size','year_built',
                                                'hoa_permonth','url']]

complete_housing_data
```

```python
complete_housing_data_clean = complete_housing_data.rename(columns={"house_id": "House ID", "property_type": "Property Type", "address": "Address", 
                                     'city':"City", 'state':"State", "zip_code":"Zip Code", 'price':"Price",'beds':"Beds",
                                      'baths':"Baths",'square_feet':"Square Feet",'lot_size':"Lot Size",
                                      'year_built':"Year Built",'hoa_permonth':"HOA Cost Per Month",'url':"URL"}, errors="raise")



                                                
complete_housing_data_clean
```

```python
training_data = housing_complete_merge[['house_id','price','beds','baths','square_feet','lot_size',
                                                'hoa_permonth', 'Summer Temp','Winter Temp','population', 'pop_cat']]

training_data
```

```python
complete_housing_data_clean.to_csv('housing_complete_data.csv', index=False)
```

```python
training_data_dropna = training_data.dropna()
```

```python
training_data_pre_hot = pd.DataFrame(training_data_dropna)
```

```python
training_data_pre_hot["lot_size_hot"] = ["1" if ele  > 0 else "0" for ele in training_data_pre_hot["lot_size"]]
training_data_pre_hot["hoa_hot"] = ["1" if ele  > 0 else "0" for ele in training_data_pre_hot["hoa_permonth"]]
# training_data_pre_hot["pop_cat_hot"] = ["0" if ele  == "small" else if "1" if  ele  == "middle" else "3" for ele in training_data_pre_hot["pop_cat"]]

```

```python
def applyFunc(s):
    if s == 'small':
        return '0'
    elif s == 'middle':
        return '1'
    elif s == 'large':
        return '2'
    return ''

training_data_pre_hot['pop_cat_hot'] = training_data_pre_hot['pop_cat'].apply(applyFunc)
```

```python
training_data_pre_hot
```

```python
training_data_pre_hot.to_csv('../housing_training_data.csv', index=False)
```
